```python
!pip install openai-whisper
!pip install ffmpeg-python
!pip install gtts
!pip install googletrans==4.0.0-rc1
import whisper
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio
speech_model = whisper.load_model("base")
audio_file = "raww.mp3"
speech_result = speech_model.transcribe(audio_file)
original_text = speech_result["text"]
print("Extracted Text:", original_text)
languages = {
    "1": ("Hindi", "hi"),
    "2": ("Telugu", "te"),
    "3": ("Tamil", "ta"),
    "4": ("French", "fr"),
    "5": ("Spanish", "es"),
    "6": ("German", "de")
}
print("\nSelect target language:")
for key, value in languages.items():
    print(f"{key}. {value[0]}")
choice = input("Enter your choice (1-6): ")
if choice not in languages:
    print("Invalid choice, defaulting to Hindi")
    selected_language = "hi"
    selected_language_name = "Hindi"
else:
    selected_language = languages[choice][1]
    selected_language_name = languages[choice][0]
print("Selected Language:", selected_language_name)
translator = Translator()
translated_result = translator.translate(original_text, dest=selected_language)
translated_text = translated_result.text
print("Translated Text:", translated_text)
tts = gTTS(text=translated_text, lang=selected_language)
output_file = "dubbed_output.mp3"
tts.save(output_file)
print("\nPlaying dubbed audio...")
Audio(output_file)
```
